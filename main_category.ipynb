```python
def filter_items(items, ctgtp_order=ctgtp_order):
    items = [item
             for item in items
             if item['categorytype'] in ctgtp_order]
    items = sorted(items,
                   key=lambda item:\
                   ctgtp_order.index(item['categorytype']))
    return items
```

```python
import numpy as np

np.random.seed(12345)
df['blank_position'] = df['items']\
    .apply(lambda items: 1+np.random.randint(0,len(items)))
```

```python
def get_item_id(set_id, items):
    return ['{}_{}'.format(set_id, item['index'])
            for item in items]
```

```python
df['item_ids'] = df.apply(lambda row:
         get_item_id(row['set_id'], row['items']),
         axis=1)
```

```python
df['question'] = df.apply(lambda row:
         row['item_ids'][:row['blank_position']-1]+\
         row['item_ids'][row['blank_position']:],
         axis=1)\
```

```python
df.apply(lambda row:
         row['items'][row['blank_position']-1]['categorytype'],
         axis=1)
```

```python
#categorytype = 'bottoms'
df_items = json_normalize(df.to_dict('records'), 'items', 'set_id')

def get_answear(name, df_items=df_items):
    ctgtp = df_items[df_items['name'] == name]['categorytype'].iloc[0]
    np.random.seed(12345)
    df_ans = df_items[(df_items['name'] != name)
            &(df_items['categorytype'] == ctgtp)].sample(3)
    return ['{}_{}'.format(row['set_id'], row['index'])
            for i, row in df_ans.iterrows()]
```

```python
df['answers'] = df.apply(lambda row:
         get_answear(row['items'][row['blank_position']-1]['name']),
         axis=1)
```

```python
#save
import json

out_file = '../data/shared/MarylandPolyvore/label/fitb_order.json'
with open(out_file, 'w') as f:
    json.dump(df[['question', 'answers', 'blank_position']].to_dict('records'),
         f, indent=4)
```
